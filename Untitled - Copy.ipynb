## Various Stages in Machine LEarning

There are a number of stages that have to be considered when carrying out a machine learning (ML) procedure - in particular, for supervised learning the following must be considered:

> Data preparation - ensuring your data is in a suitable format to apply the ML algorithm.

> Choice of ML algorithm.

```python
from sklearn.pipeline import Pipeline
```

We can use pipelines with GridSearchGV. GridSearchCV lets you combine an estimator with a grid search preamble to tune hyper-parameters. The method picks the optimal parameter from the grid search and uses it with the estimator selected by the user.

However in the dictionary part of the param_grid argument the parameters must be referred to in the form < name>__ < parameter> where < name> is the block name in the pipeline, and < parameter> is the tuning parameter name – note there are two underline characters. For example NNreg__n_neighbors refers to the number of nearest neighbours parameter in the NNreg block.

```python
from sklearn.preprocessing import StandardScaler
# For standard scaling a function called StandardScaler is used. This does not carry out the scaling itself, 
# but produces a Python object that will do the scaling for a specific variable, or set of variables.
from sklearn.neighbors import KNeighborsRegressor as NN
# We need to import the KNeighborsRegressor function from the sub-package sklearn.neighbors. For brevity here it is 
# imported as NN.Rather like the StandardScaler function earlier, NN is actually a function to produce nearest neighbour 
# regression objects.
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, make_scorer
mae = make_scorer(mean_absolute_error, greater_is_better=False)
# The greater_is_better=False argument is used to specify that in this case lower scores correspond to better models.
x_scaler = StandardScaler()

pipe = Pipeline([('zscores',StandardScaler()),('NNreg',NN())])

opt_nn2 = GridSearchCV(
    estimator = pipe,
    scoring = mae,
    param_grid = {
        'NNreg__n_neighbors':range(1,35),
        'NNreg__weights':['uniform','distance'],
        'NNreg__p':[1,2]})
price = hp['price']/1000.0
opt_nn2.fit(hp[['east','north','fl_area']],price)
```

## Visualising the Data

After reading the data next step is to visualise the data based on the conditions that are:-

> A 3-D plot of the house price for average floor area.

> A 3-D plot of the house price for floor areas of 75 square metres.

> A 3-D plot of the house price for floor areas of 125 square meters.

To get a broader idea of geographical trends of the house prices based on the surface for London area we have to create a grid of easting and northing points to cover London, and obtaining an estimate of house price at each grid point. This would require floor area in addition to easting and northing so firstly we will look at predicted house price ata given location assuming a house has the average floor area.

A 3D surface plot of the predictions for each of the specified house size was created by firstly generating a 100x100 grid covering the range of easting and northing values, and then calculating the predicted house price at each grid location for the specified floor area. To cover the area in the data eastings should range from 505000 to 555800, and northings from 158400 to 199900. For the geographical grid we will use meshgrid function of NumPy

```python
east_mesh, north_mesh = np.meshgrid(
    np.linspace(505000,555800,100),
    np.linspace(158400,199900,100))
```

The numpy.meshgrid function is used to create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing. In this case easting and northing are the two one dimensional arrays which needs to be converted to the regular grid for the study area.

The NumPy linspace function (sometimes called np. linspace) is a tool in Python for creating numeric sequences. It's somewhat similar to the NumPy arange function, in that it creates sequences of evenly spaced numbers structured as a NumPy array.In this case it is creating a sequence from 505000 - 555800 on the sacle of 100 for east_mesh and 158400 - 199900 on the scale of 100 for north_mesh. 

Next, a third grid of floor area values is needed. Earlier we assumed that the average level should be used, the following technique is used to generate the variable fl_mesh.

```python
fl_mesh = np.zeros_like(east_mesh)
```

Here we are creating a array of the same size of the east_mesh as we have created above, because of zeros_like function  of NumPy that we have used here the elements of this array will contain zeros only. We have created this array just because to create an array of the correct size. After creating this array we can assign every element of this array with the average floor size.

```python
fl_mesh[:,:] = np.mean(hp['fl_area'])
```

This piece of code will convert fl_area of house price to its mean and add it to our array with all zeros mean function of NumPy will help in this.

After this we have all the three predictors that we require to plot a 3D plot but in the form 2D arrays.

However, opt_nn2 that we have defined for tunning the data doesn’t expect 2D arrays. For the prediction, each variable should be ‘unravelled’ into single dimensional array. To unravel all three variables and put them into the format that the predict method expects, the expression below is used:

```python
grid_predictor_vars = np.array([east_mesh.ravel(),
    north_mesh.ravel(),fl_mesh.ravel()]).T 
```

The ravel method which is used here is used so that the values are just listed as though reading through the matrix as a series running left to right from top to bottom.

```python
hp_pred = opt_nn2.predict(grid_predictor_vars)
```

Now the only issue is that hp_pred contains the house prices, but in an unravelled form. To plot the surface, it needs to be back in grid format. This can be achieved with the reshape method. 

The reshape() function is used to give a new shape to an array without changing its data. Array to be reshaped. The new shape should be compatible with the original shape. If an integer, then the result will be a 1-D array of that length.

This takes a tuple of dimensions (for example the number of rows and columns needed) and re-shapes an array into this form. It effectively re-arranges an unravelled array back to its 2D form. It can be used in conjuction with the shape attribute of an array to re-shape one array back to the shape of another one. Here hp_pred is re-shaped to have the same dimensions as east_mesh, and stored in a new array called hp_mesh.

```python
hp_mesh = hp_pred.reshape(east_mesh.shape)
```

Once this is done, it is possible to apply the plotting approaches to create a 3D plot of predicted house prices for a house with mean or average floor area.

> ### 3D Plot of the house price for average floor area:-

```python
import pylab as pl
from mpl_toolkits.mplot3d import Axes3D

fig = pl.figure()
ax = Axes3D(fig)
ax.plot_surface(east_mesh, north_mesh, hp_mesh, rstride=1, cstride=1, cmap='YlOrBr',lw=0.01)
ax.set_xlabel('Easting')
ax.set_ylabel('Northing')
ax.set_zlabel('Price at Mean Floor Area')
ax.set_title('House Price For Average Floor Area')
pl.show()
```

mpl_toolkits.mplot3d is used to generate the 3D plot. From the plot we can conclude that for average floor area the heighest price is around 175k pounds.

As we have to create a 3D plot with floor area 75 and 125 we will create a function which we will call with different floor areas based on our requirement.

```python
def surf3d(pipe_model,fl_area):
    east_mesh, north_mesh = np.meshgrid(
        np.linspace(505000,555800,100),
        np.linspace(158400,199900,100))
    fl_mesh = np.zeros_like(east_mesh)
    fl_mesh[:,:] = fl_area
    grid_predictor_vars = np.array([east_mesh.ravel(),
        north_mesh.ravel(),fl_mesh.ravel()]).T
    hp_pred = pipe_model.predict(grid_predictor_vars)
    hp_mesh = hp_pred.reshape(east_mesh.shape)
    fig = pl.figure()
    ax = Axes3D(fig)
    ax.plot_surface(east_mesh, north_mesh, hp_mesh, 
        rstride=1, cstride=1, cmap='YlOrBr',lw=0.01)
    ax.set_xlabel('Easting')
    ax.set_ylabel('Northing')
    ax.set_zlabel('Price at Mean Floor Area')
    pl.show()
    return
```

> ### 3D Plot with floor area 75 sq metres.

```python
pl.close()
surf3d(opt_nn2,75.0)
pl.show()
```

> ### 3D Plot with Floor Area 125 Sq metres.

```python
pl.close()
surf3d(opt_nn2,125.0)
pl.show()
```

Each of the 3D plot shows the predicted house value for given floor area at each easting & northing location in London. The darker color represent high price locations whereas lighter color represent lower prices for all the 3 floor areas i.e Average floor area, 75 sq mtrs and 125 sq mtrs.
